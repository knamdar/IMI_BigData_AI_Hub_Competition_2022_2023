```python
test= test.drop(['Unnamed: 0'],axis =1)[:100]
```

```python
y_test= y_test.drop(['Unnamed: 0'],axis=1)[:100]
```

```python
#y_test.Risk.tolist()
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=10)
```

```python
clf.fit(test,y_test)
```

```python
viz= dtreeviz(clf, 
               x_data=test,
               y_data=y_test.Risk,
               target_name='Classes',
               feature_names=list(test.columns), 
               class_names=[0,1,2], 
               title="Decision Tree - including clusters")
```
